```python
wL = 6.7*10**12
wT = 4.3 * 10 ** 12
```

```python
hbar = 1.05*10**-34

meVToJ = 1.6*10**-22
def g(w, E):
    return 1/w * (np.pi * (hbar**2-E**2/4 * w**-2))**-.5
```

```python
E = np.linspace(0, 12*meVToJ, 30)
EMeV = np.linspace(0, 12, 30)
```

```python
gT = g(wT, E) 
```

```python
gL = g(wL, E) 
```

```python
gTot = [(gT[i] + gL[i] * 2 )* meVToJ for i in range(len(gT))]
```

```python
plt.plot(gTot, EMeV)
plt.xlabel('density of states 1/meV')
plt.ylabel('energy meV')
plt.title('Density of States graph')
```

```python
def evenOdd(data):
    even = []
    odd = []
    for i in range(len(data['x'])):
        if i % 2 == 0:
            even += [list(data['x'])[i]]
        elif i % 2 == 1:
            odd += [list(data['x'])[i]]
    diff = [even[i] - odd[i] for i in range(len(odd))]
    return even, odd, diff
            
```

```python
def splitByCurrent(data):
    iData = list(data['i'])
    xData = list(data['x'])
    tData = list(data['t'])
    
    posCurrX = []
    negCurrX = []
    
    posCurrT = []
    negCurrT = []
    

    for j in range(len(iData)):
        if iData[j] > 0:
            posCurrX  += [xData[j]]
            posCurrT += [tData[j]]
        else:
            negCurrX += [xData[j]]
            negCurrT += [tData[j]]

    return posCurrX, negCurrX, posCurrT, negCurrT

def diffTwo(data):
    posCurrX, negCurrX, posCurrT, negCurrT = splitByCurrent(data)
    if len(posCurrX) > len(negCurrX):
        posCurrX = posCurrX[1:]
        posCurrT = posCurrT[1:]
    elif len(posCurrX) < len(negCurrX):
        negCurrX = negCurrX[1:]
        negCurrT = negCurrT[1:]
    
    diffX = [posCurrX[i] - negCurrX[i] for i in range(len(negCurrX))]
    offX = [(posCurrX[i] + negCurrX[i])/2 for i in range(len(negCurrX))]
    avgT = [(posCurrT[i] + negCurrT[i])/2 for i in range(len(negCurrX))]
    return avgT, diffX, offX
```
