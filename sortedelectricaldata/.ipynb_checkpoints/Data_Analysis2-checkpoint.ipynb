```python
data = pd.read_csv("datalogging/Data/HallEffectDCBias/hallDCBiasSMU1681453699.txt")

def avgHallSignal(data, dcList):
    #dcList = np.linspace(-200, 200, 21)
    dataP = data[data['i'] == 3.2]
    dataN = data[data['i'] == -3.2]
    avgP = []
    avgN = []

    for dc in dcList:
        dataDC = dataP[abs(dataP['dc'] - dc) < .1]
        avgX = np.mean(dataDC['x'])
        avgP += [avgX]
                  
    for dc in dcList:
        dataDC = dataN[abs(dataN['dc'] - dc) < .1]
        avgX = np.mean(dataDC['x'])
        avgN += [avgX]
    
    avgDiff = [avgP[i]-avgN[i] for i in range(len(avgP))]
    return avgDiff
```

```python
def avgHallSignal(data):
    dcList = data['trueGateDC'].unique()
    dataP = data[data['i'] == 3.2]
    dataN = data[data['i'] == -3.2]
    avgP = []
    avgN = []

    for dc in dcList:
        dataDC = dataP[abs(dataP['trueGateDC'] - dc) < .1]
        avgX = np.mean(dataDC['x'])
        avgP += [avgX]
                  
    for dc in dcList:
        dataDC = dataN[abs(dataN['trueGateDC'] - dc) < .1]
        avgX = np.mean(dataDC['x'])
        #linOffset = (list(dataDC['x'])[-1]-list(dataDC['x'])[0])/(len(dataDC)*2)
        #linOffset = 0
        linOffset = 0
        avgN += [avgX-linOffset]
    
    avgDiff = [avgP[i]-avgN[i] for i in range(len(avgP))]
    return avgDiff
```
